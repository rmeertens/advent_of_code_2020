```python
lines = """F10
N3
F7
R90
F11"""
lines = [l for l in lines.split("\n")]
lines = [l for l in open("inputs/day12.txt")]
```

```python
# given in X, Y coordinates, incrementing with 1 turns right
directions = [(1, 0), (0, -1), (-1, 0), (0, 1)] 

position = (0,0)
facing = 0

for instr in lines: 
    letter, action = instr[0], int(instr[1:])
#     print(position, facing)
    for _ in range(action):
        if letter == "N":
            position = (position[0], position[1]+1)
        elif letter == "E": 
            position = (position[0]+1, position[1])
        elif letter == "S": 
            position = (position[0], position[1]-1)
        elif letter == "W": 
            position = (position[0]-1, position[1])
        elif letter == "F": 
            position = (position[0]+directions[facing][0], 
                        position[1]+directions[facing][1])
        elif letter == "L":
            facing = (facing-action//90) %len(directions)
            break
        elif letter == "R":
            facing = (facing+action//90) %len(directions)
            break
print("Part 1", abs(position[0])+abs(position[1]))
```
