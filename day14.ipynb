```python
from collections import defaultdict
mask = ''
mem = defaultdict(int)
for line in inputs: 
    if 'mask' in line: 
        mask = line.split(' ')[-1]
        print(mask)
    else:
        index = line[4:].split(']')[0]
        value = line.split(' ')[-1]
        bin_res = [x for x in "{0:b}".format(int(value)).zfill(36)]
        print(''.join(bin_res))
#         print(index, value, bin_res)        
        for i, v in enumerate(mask):
            if v != 'X':
                bin_res[i] = v
        mem[index] = int(''.join(bin_res),2)
#         print(''.join(bin_res), int(''.join(bin_res),2))
sum(mem.values())
```

```python
def yield_combinations(mask):
    index = mask.find('X')
    if index >= 0:
        print(mask[index+1:])
        for end in yield_combinations(mask[index+1:]):
            yield mask[:index]+'0'+end
            yield mask[:index]+'1'+end
    else: 
        yield mask
for c in yield_combinations('000000000000000000000000000000X1001X'):
    print(c)
```

```python
from collections import defaultdict
mask = ''
mem = defaultdict(int)
for line in inputs: 
    print('---')
    if 'mask' in line: 
        mask = line.split(' ')[-1]
        print(mask)
    else:
        address = line[4:].split(']')[0]
        value = line.split(' ')[-1]
        bin_address = [x for x in "{0:b}".format(int(address)).zfill(36)]
        print(''.join(bin_address))
#         print(index, value, bin_res)        
        for i, v in enumerate(mask):
            if v != '0':
                bin_address[i] = v
        print(''.join(bin_address))
        
        for c in yield_combinations(''.join(bin_address)):
            
            mem[int(''.join(c),2)] = int(value)
#         print(''.join(bin_res), int(''.join(bin_res),2))
sum(mem.values())
```
